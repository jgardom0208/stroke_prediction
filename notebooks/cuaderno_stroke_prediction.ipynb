## Predicción de ictus mediante inteligencia artificial

He elegido este dataset sobre la predicción de ictus porque según la Organización Mundial de la Salud esta enfermedad es la segunda causa de muerte en el planeta y representa el 11 por ciento de los fallecimientos totales.

El objetivo principal de mi trabajo es usar esta información para crear un modelo de inteligencia artificial capaz de identificar a pacientes con riesgo de sufrir un infarto cerebral.

### Descripción del conjunto de datos

El conjunto de datos incluye información de los pacientes repartida en las siguientes categorías:

#### Información personal
Incluye datos como el género, la edad, el estado civil, el tipo de trabajo y si la persona vive en una zona rural o urbana.

#### Historial clínico
Sirve para saber si el paciente padece hipertensión o alguna enfermedad del corazón.

#### Datos de salud actuales
Aquí se incluyen el nivel de glucosa en sangre, el índice de masa corporal y si la persona fuma o ha fumado anteriormente.

#### Resultado
La variable objetivo se llama stroke y marca con un 1 si el paciente ha tenido un ictus o con un 0 si no lo ha tenido.

Este dataset permite analizar distintos factores de riesgo y es adecuado para entrenar un modelo predictivo aplicado al ámbito de la salud.

# 1. Preparacion de los datos

```python
import pandas as pd
```

```python
#Cargamos el dataset
df = pd.read_csv('./datasets/healthcare-dataset-stroke-data.csv')
# y vemos su tamaño y sus primeras filas
print(len(df))
df.head().T
```

```python
df.dtypes
```

Como todos los tipos de datos estan con su tipo no haria falta hacer ningun cambio 

```python
# Información básica del DataFrame para ver tipos de datos y valores nulos
df.info()
```

```python
# Como podemos ver en la columna 'bmi' hay valores nulos
df['bmi'].isnull().sum()
# Vemos las filas con valores nulos en la columna 'bmi'
df[df['bmi'].isnull()]
```

```python
# y las remplazamos por la mediana de la columna para no afectar la distribución
df['bmi'] = df['bmi'].fillna(df['bmi'].median())
# Verificamos que ya no haya valores nulos en la columna 'bmi'
df['bmi'].isnull().sum()
```

```python
# Ahora vamos a normalizar los valores de texto
# normalizar los valores de texto de los nombres de las columnas
df.columns = df.columns.str.lower().str.replace(' ', '_')
# Normalizamos los valores de texto en todo el dataset
replacer = lambda s: s.lower().str.replace(' ', '_')
for col in df.select_dtypes(include='object').columns:
    df[col] = replacer(df[col].str)
df.head().T
```

```python
#Vamos a ver el numero de valores unicos por columna
categorical = [
    'gender', 
    'ever_married', 
    'work_type', 
    'residence_type', 
    'smoking_status', 
    'hypertension', 
    'heart_disease'
]
numerical = [
    'age', 
    'avg_glucose_level', 
    'bmi'
]
df[categorical].nunique()
```

```python
#Ahora con todo preparado vamos a separar los datos
from sklearn.model_selection import train_test_split
#Dividimos en entrenamiento y testn es train full ya que despues se sacara validacion
df_train_full , df_test = train_test_split(df, test_size=0.2, random_state=1)

# Dividimos en entrenamiento y validacion
df_train , df_val = train_test_split(df_train_full, test_size=0.33, random_state=1)

# Guardammos las etiquetas de los ejemplos en una variable
y_train = df_train.stroke.values
y_val = df_val.stroke.values
y_test = df_test.stroke.values

# Eliminamos la columna de etiquetas de los datos
del df_train['stroke']
del df_val['stroke']
del df_test['stroke']
```

```python
# Vemos el tamaño de las particiones
print(len(df_train) ,len(y_train), 'train examples')
print(len(df_val),len(y_val), 'validation examples')
print(len(df_test), len(y_test), 'test examples')
```

# 2. Análisis de importancia de prioridades


```python
# Calculamos la media global de las personas que han tenido un ictus
global_mean = df_train_full.stroke.mean()
print(f'Media global de ictus: {round(global_mean, 3)}')
```

Vemos que esuna enfermedad poco frecuente , lo que hace que el reto de la IA sea encontrar los patrones específicos que distinguen a ese pequeño grupo del resto ya que solo un 4,6% padecen de esta. Tambien como ya sabemos que hay pocos implica que una precisión alta puede ser engañosa porque el modelo podría acertar casi siempre simplemente prediciendo que nadie tendrá un ictus

```python
#Vamos a ver si el genero influye en la media de ictus
female_mean = df_train_full[df_train_full.gender == 'female'].stroke.mean()
print(f'Riesgo en mujeres: {round(female_mean, 3)}')

male_mean = df_train_full[df_train_full.gender == 'male'].stroke.mean()
print(f'Riesgo en hombres: {round(male_mean, 3)}')
```

En mi análisis el riesgo de sufrir un ictus es muy similar entre ambos sexos aunque las mujeres presentan una probabilidad ligeramente superior del 4.8% frente al 4.4% de los hombres. Esto indica que el género no es el factor más determinante de mi modelo pero sigue siendo una variable importante para realizar una predicción precisa.

```python
# Vamos a ver tambien si el estado de fumador influye en la media de ictus
never_mean = df_train_full[df_train_full.smoking_status == 'never_smoked'].stroke.mean()
print(f'Riesgo nunca han fumado: {round(never_mean, 3)}')

smokes_mean = df_train_full[df_train_full.smoking_status == 'smokes'].stroke.mean()
print(f'Riesgo fumadores actuales: {round(smokes_mean, 3)}')

formerly_mean = df_train_full[df_train_full.smoking_status == 'formerly_smoked'].stroke.mean()
print(f'Riesgo ex-fumadores: {round(formerly_mean, 3)}')
```

Sobre el tabaquismo destaca que los ex-fumadores tienen el riesgo más alto con un 7.2% superando notablemente a quienes fuman actualmente o nunca lo han hecho. Estos resultados sugieren que el daño acumulado o el historial previo de tabaquismo es un factor de riesgo mucho más crítico en mi dataset que el hábito actual.

```python
from sklearn.metrics import mutual_info_score
# Calculo de la informacion mutua para cada variable categorica 
# la correlaccion que tiene la variable objetivo con cada variable 
# , solo para categoricas , si es 0 no tiene relacion y mientras mas alto mas relacion 
calculate_mi = lambda col: mutual_info_score(col, df_train_full.stroke)

df_mi = df_train_full[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')
df_mi
```

El análisis de la información mutua nos muestra que la mas determinante es el tipo de trabajo , seguido de la hipertensión y de los infartos , el genero o la residencia son variables que no aportan mucho a si va a sufrir un ictus o no .

```python
# Calculo de la correlacion para las variables numericas (coeficiente de correlacion de Pearson)
print(df_train_full[numerical].corrwith(df_train_full.stroke))
```

Este análisis de correlación de Pearson muestra que la edad es el factor numérico más influyente ya que presenta la relación positiva más fuerte con el riesgo de ictus. El nivel de glucosa también aporta información relevante para el modelo mientras que el índice de masa corporal tiene el impacto más bajo de los tres valores analizados. Esto confirma que en mi dataset el riesgo de sufrir un infarto aumenta principalmente a medida que los pacientes envejecen y tienen niveles altos de azúcar.

```python
# Riesgo de ictus según rangos de EDAD
print("Riesgo por edad (Jóvenes <= 30):")
print(round(df_train_full[df_train_full.age <= 30].stroke.mean(), 3))

print("Riesgo por edad (Adultos 31-60):")
print(round(df_train_full[(df_train_full.age > 30) & (df_train_full.age <= 60)].stroke.mean(), 3))

print("Riesgo por edad (Mayores > 60):")
print(round(df_train_full[df_train_full.age > 60].stroke.mean(), 3))

# Riesgo de ictus según rangos de GLUCOSA
print("\nRiesgo por glucosa (Normal <= 100):")
print(round(df_train_full[df_train_full.avg_glucose_level <= 100].stroke.mean(), 3))

print("Riesgo por glucosa (Moderada 101-150):")
print(round(df_train_full[(df_train_full.avg_glucose_level > 100) & (df_train_full.avg_glucose_level <= 150)].stroke.mean(), 3))

print("Riesgo por glucosa (Alta > 150):")
print(round(df_train_full[df_train_full.avg_glucose_level > 150].stroke.mean(), 3))
```

Esto confirma lo visto en la correlaccion , la edad y el azucar en sangre son los predictores mas potentes 

# 3. Ingenieria de prioridades

```python
# lo pasamos a un diccionario y mostramos el primero
train_dict = df_train[categorical + numerical].to_dict(orient='records')
val_dict = df_val[categorical + numerical].to_dict(orient='records')
dict(sorted(train_dict[0].items()))
```

```python
from sklearn.feature_extraction import DictVectorizer
# Creamos el vectorizador y lo ajustamos a los datos de entrenamiento , esta tecnica es para hacer el one hot encoding
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
# transformamos los datos a formato numérico
x_train = dv.transform(train_dict)
x_val = dv.transform(val_dict)
x_train[0]
x_val[0]
```

```python
# Podemos ver los nombres de las nuevas columnas generadas
dv.get_feature_names_out()
```

# 4. Entrenamiento de los modelos 

```python
# Importamos los modelos y la matriz de confusión
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
```

## 4.1 Regresión logística

```python
model_lr = LogisticRegression(solver='liblinear', random_state=1)
model_lr.fit(x_train, y_train)
```

## 4.2 SVM

```python
# Probamos inicialmente con un kernel lineal
model_svm = SVC(kernel='linear', probability=True, random_state=1)
model_svm.fit(x_train, y_train)
```

## 4.3 Arbol de decición

```python
model_dt = DecisionTreeClassifier(max_depth=5, random_state=1)
model_dt.fit(x_train, y_train)
```

## Evaluación de modelos

```python
def evaluar_modelo(nombre, modelo, x, y_real):
    # Predicción de etiquetas
    y_pred = modelo.predict(x)
    
    # Cálculo manual de la precisión (sin bibliotecas externas) [cite: 41]
    accuracy = (y_real == y_pred).mean()
    
    # Matriz de confusión [cite: 41]
    conf_mat = confusion_matrix(y_real, y_pred)
    
    print(f"{nombre}")
    print(f"Precisión (manual): {round(accuracy, 3)}")
    print("Matriz de Confusión:")
    print(conf_mat)

evaluar_modelo("Regresión Logística", model_lr, x_val, y_val)
evaluar_modelo("SVM", model_svm, x_val, y_val)
evaluar_modelo("Árbol de Decisión", model_dt, x_val, y_val)
```

Los modelos base sufren de un gran desequilibrio. La Regresión Logística y la SVM inicial predecían siempre "no ictus", logrando un 95.7% de precisión pero siendo inútiles para detectar enfermos . Vamos a usar el gridsearh para encontrar algun modelo que logre conseguir mejores serultados

```python
from sklearn.model_selection import GridSearchCV

# Definimos los parámetros a probar para la SVM
# añadimos el class_weight para balancear las clases
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'class_weight': ['balanced'] 
}
```

```python
# Configuramos la búsqueda. Usamos scoring='f1' o 'recall' para priorizar detectar los ictus
grid_svm = GridSearchCV(SVC(probability=True), param_grid, cv=3, scoring='f1')
grid_svm.fit(x_train, y_train)
```

```python
# Mostramos la mejor configuración encontrada
print(f"Mejores parámetros para SVM: {grid_svm.best_params_}")
```

```python
# Evaluamos el mejor modelo encontrado
best_svm = grid_svm.best_estimator_
y_pred_best = best_svm.predict(x_val)

print(f"Nueva Precisión SVM: {round((y_val == y_pred_best).mean(), 3)}")
print("Nueva Matriz de Confusión:")
print(confusion_matrix(y_val, y_pred_best))
```

Al aplicar GridSearch y pesos balanceados corregimos el subajuste inicial. Aunque la precisión bajó al 72.4% el modelo ahora detecta 43 pacientes en riesgo y reduce drásticamente los falsos negativos. Esta configuración es la mejor para el ámbito médico porque prioriza la detección real de la enfermedad sobre la simple precisión estadística. Para asegurarnos de tener el mejor modelo vamos a hacerlo tambien con la regresion logística y el arbol de decisión

```python
# Definimos los parámetros a probar para la Regresión Logística
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear'],
    'class_weight': ['balanced']
}

grid_lr = GridSearchCV(LogisticRegression(max_iter=1000), param_grid_lr, cv=3, scoring='f1')
grid_lr.fit(x_train, y_train)

best_lr = grid_lr.best_estimator_
print(f"Mejores parámetros Regresión Logística: {grid_lr.best_params_}")

# Evaluamos el mejor modelo encontrado
y_pred_best_lr = best_lr.predict(x_val)
print(f"Nueva Precisión Regresión Logística: {round((y_val == y_pred_best_lr).mean(), 3)}")
print("Nueva Matriz de Confusión:")
print(confusion_matrix(y_val, y_pred_best_lr))
```

```python
# Definimos los parámetros a probar para el Árbol de Decisión
param_grid_dt = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy'],
    'class_weight': ['balanced']
}

grid_dt = GridSearchCV(DecisionTreeClassifier(), param_grid_dt, cv=3, scoring='f1')
grid_dt.fit(x_train, y_train)

best_dt = grid_dt.best_estimator_
print(f"Mejores parámetros Árbol de Decisión: {grid_dt.best_params_}")

# Evaluamos el mejor modelo encontrado
y_pred_best_dt = best_dt.predict(x_val)
print(f"Nueva Precisión Árbol de Decisión: {round((y_val == y_pred_best_dt).mean(), 3)}")
print("Nueva Matriz de Confusión:")
print(confusion_matrix(y_val, y_pred_best_dt))
```

Ambos modelos logran equilibrar la detección, identificando 42 y 37 casos de ictus respectivamente. El Árbol de Decisión destaca por ofrecer la precisión más alta y el menor número de falsas alarmas, aunque es el que más casos reales deja sin detectar entre los tres modelos analizados.

El SVM es el modelo superior al detectar 43 casos positivos y minimizar los fallos a solo 15. Aunque su precisión sea ligeramente menor prioriza la seguridad del paciente al reducir drásticamente los falsos negativos.
